```python
import pandas as pd
import imdb
```

```python
mi = imdb.Cinemagoer()
```

```python
df = pd.read_excel(r"D:\Naveen\Entertainer PowerBi\Entertainer - Basic Info.xlsx", engine="openpyxl", usecols=[0])
```

```python
df.head()
```

```python
actor = []
movie_title = []
year = []
```

```python
for i in range(len(df)):
    ent_name = df.Entertainer[i]
    leen = len(movie_title)
    print("Start-->", ent_name)
    persons = mi.search_person(ent_name)
    try:
        per = mi.get_person(persons[0].personID)
        if 'filmography' in per.keys():
            ma_keys = per["filmography"].keys()
            if "actor" in ma_keys:
                act = per["filmography"]["actor"]
                
                for j in range(len(act)):
                    movi = mi.get_movie(act[j].movieID)
                    k = movi.keys()
                    actor.append(ent_name)
                    if "title" in k:
                        movie_title.append(movi["title"])
                    else:
                        movie_title.append("NA")
                    
                    if "year" in k:
                        year.append(movi["year"])
                    else:
                        year.append("NA")
                        
            elif "actress" in  ma_keys:
                act = per["filmography"]["actress"]
                
                for j in range(len(act)):
                    movi = mi.get_movie(act[j].movieID)
                    k = movi.keys()
                    actor.append(ent_name)
                    if "title" in k:
                        movie_title.append(movi["title"])
                    else:
                        movie_title.append("NA")
                    
                    if "year" in k:
                        year.append(movi["year"])
                    else:
                        year.append("NA")
            else:
                actor.append(ent_name)
                movie_title.append("NA")
                year.append("NA")
        else:
            actor.append(ent_name)
            movie_title.append("NA")
            year.append("NA")
    except:
        actor.append(ent_name)
        movie_title.append("NA")
        year.append("NA")  
    print("Done-->", ent_name, " Added--> ", len(movie_title)-leen, " Movies")
    print("   ")            
```
